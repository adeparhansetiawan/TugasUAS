```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Salary_Data.csv')
df
```

```python
df.describe()
```

```python
x = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(x,y)

prediksi = model.predict(x)
print(f"Hasil Prediksi: {prediksi}")
```

```python
pred = model.predict(x)

plt.figure(figsize=(10,5))
sns.scatterplot(x=df['YearsExperience'], y=df['Salary'], label='Data Asli')
plt.plot(df['YearsExperience'], pred, color='black', label='Garis Regresi')

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title("Prediksi Profit Berdasarkan YearsExperience")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
